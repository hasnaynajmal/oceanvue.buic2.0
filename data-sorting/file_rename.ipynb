```python
import os

def rename_directories(folder_path):
    for dir_name in os.listdir(folder_path):
        if os.path.isdir(os.path.join(folder_path, dir_name)):
            new_name = dir_name[-1]  # Keep the last character of the directory name
            new_path = os.path.join(folder_path, new_name)
            os.rename(os.path.join(folder_path, dir_name), new_path)
            print(f"Renamed {dir_name} to {new_name}")

# Replace 'folder_path' with the path to the folder containing the directories
folder_path = 'C:/Users/oceanvue/Desktop/rename'
rename_directories(folder_path)
```

```python
import os

def rename_directories(folder_path):
    for dir_name in os.listdir(folder_path):
        if os.path.isdir(os.path.join(folder_path, dir_name)):
            new_name = dir_name.split('-')[-1]  # Get characters after the last '-'
            new_path = os.path.join(folder_path, new_name)
            os.rename(os.path.join(folder_path, dir_name), new_path)
            print(f"Renamed {dir_name} to {new_name}")

# Replace 'folder_path' with the path to the folder containing the directories
folder_path = 'C:/Users/oceanvue/Desktop/DeepShip Dataset/Tanker'
rename_directories(folder_path)
```

```python
import os

def rename_files_in_folders(folder_path):
    for dir_name in os.listdir(folder_path):
        dir_path = os.path.join(folder_path, dir_name)
        if os.path.isdir(dir_path):
            for file_name in os.listdir(dir_path):
                file_path = os.path.join(dir_path, file_name)
                new_file_name = dir_name + os.path.splitext(file_name)[1]
                new_file_path = os.path.join(dir_path, new_file_name)
                os.rename(file_path, new_file_path)
                print(f"Renamed {file_name} to {new_file_name}")

# Replace 'folder_path' with the path to the folder containing the subfolders with files
folder_path = 'C:/Users/oceanvue/Desktop/DeepShip Dataset/Tanker'
rename_files_in_folders(folder_path)
```
