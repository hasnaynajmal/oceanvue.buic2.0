```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
# Define the path to your audio file
audio_file_path = 'D:/oceanvue.fyp/DeepShip Dataset/Passenger/2/2.wav'
```

```python
# Load the audio file using librosa
y, sr = librosa.load(audio_file_path)
```

```python
# Calculate the Mel spectrogram
mel_spectrogram = librosa.feature.melspectrogram(y=y, sr=sr)

# Convert the power spectrogram to dB scale (decibels)
mel_spectrogram_db = librosa.power_to_db(mel_spectrogram, ref=np.max)
```

```python
mel_spectrogram_db
```

```python
# Display the Mel spectrogram
plt.figure(figsize=(5, 3))
librosa.display.specshow(mel_spectrogram_db, x_axis='time', y_axis='mel', sr=sr, fmax=8000) # Adjust fmax as needed
```
